```python
# =============================================================================
#  PHASE 2 — DATA CLEANING
#  Project : Crop Yield Prediction Using Environmental and Nutrient Factors
#  Course  : Data Warehousing and Data Mining (DWDM)
# =============================================================================

import pandas as pd
import numpy as np

df = pd.read_csv("crop_yield_raw_dataset.csv")

print(f"   Rows    : {df.shape[0]:,}")
print(f"   Columns : {df.shape[1]}")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 0 — INITIAL DATA QUALITY REPORT  (show problems BEFORE cleaning)
# ─────────────────────────────────────────────────────────────────────────────

df.shape,df.duplicated().sum()
```

```python
missing = df.isnull().sum()
miss_df = pd.DataFrame({"Missing Count": missing})
miss_df = miss_df[miss_df["Missing Count"] > 0]
print(miss_df.to_string())
```

```python
print(df.dtypes.to_string())
# df.describe().round(2).to_string()
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 2 — STRUCTURAL AND LOGICAL FIXES
# ─────────────────────────────────────────────────────────────────────────────

# ── 2A : Fix negative elevation values ──────────────────────────────────────
# Problem : 5,504 rows have negative Elevation_m (e.g. -133.5 m)
# Reason  : GPS data entry error — no Indian farmland is below sea level
# Fix     : Convert to absolute value

neg_count = (df["Elevation_m"] < 0).sum()
df["Elevation_m"] = df["Elevation_m"].abs()

neg_count,df['Elevation_m'].min()
```

```python
# ── 2B : Convert zero micronutrients to NaN ──────────────────────────────────
# Problem : Iron, Zinc, Sulfur columns have zeros
# Reason  : When soil test was not done, 0 was entered instead of leaving blank
#           A true zero iron/zinc content is agronomically impossible
# Fix     : Replace 0 with NaN — will be imputed in Step 3

zero_fe = (df["Iron_ppm"]    == 0).sum()
zero_zn = (df["Zinc_ppm"]    == 0).sum()
zero_s  = (df["Sulfur_kgha"] == 0).sum()

df.loc[df["Iron_ppm"]    == 0, "Iron_ppm"]    = np.nan
df.loc[df["Zinc_ppm"]    == 0, "Zinc_ppm"]    = np.nan
df.loc[df["Sulfur_kgha"] == 0, "Sulfur_kgha"] = np.nan

zero_fe,zero_zn,zero_s
```

```python
# ── 2C : Fill categorical NaN with 'None' (not truly missing) ───────────────
# Problem : Pest_Incidence and Disease_Incidence have ~55,000 NaN each (~25%)
# Reason  : When no pest/disease was observed, the field was left blank
#           NaN here means "None observed", NOT missing data
# Fix     : Fill NaN with the string 'None'

pest_nan = df["Pest_Incidence"].isna().sum()
dis_nan  = df["Disease_Incidence"].isna().sum()

df["Pest_Incidence"]  .fillna("None")
df["Disease_Incidence"].fillna("None")

print(f"{pest_nan:,}")
print(f"{dis_nan:,}")
print(f"{df['Pest_Incidence'].value_counts().to_dict()}")
```

```python
# ── 2D : Fix Yield_Category labels (CRITICAL FIX) ───────────────────────────
# Problem : Raw data had 93.8% records labelled 'Low', only 9 as 'High'
# Reason  : The original code used fixed brackets for ALL crops together
#           Sugarcane (20,000–80,000 kg/ha) and Chickpea (300–1,300 kg/ha)
#           cannot share the same yield category scale
# Fix     : Recalculate Low/Medium/High using the 33rd and 66th percentile
#           WITHIN each crop separately
 
print(f"\n[2D] Yield_Category Correction (Per-Crop Percentile)")
print(f"     {df['Yield_Category'].value_counts().to_dict()}")
 
df["Yield_Category"] = "Medium"
 
for crop in df["Crop_Type"].unique():
    mask   = df["Crop_Type"] == crop
    yields = df.loc[mask, "Crop_Yield_kg_ha"]
    p33    = yields.quantile(0.33)
    p66    = yields.quantile(0.66)
 
    df.loc[mask & (yields <= p33), "Yield_Category"] = "Low"
    df.loc[mask & (yields >  p66), "Yield_Category"] = "High"
 
    print(f"     {crop:<12} → Low ≤ {p33:.0f}  |  Medium  |  High > {p66:.0f} kg/ha")
 
print(f"     {df['Yield_Category'].value_counts().to_dict()}")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 3 — MISSING VALUE IMPUTATION
# ─────────────────────────────────────────────────────────────────────────────

miss_now = df.isnull().sum()
print(miss_now[miss_now > 0].to_string())

# ── 3A : NDVI — median per Crop_Type ────────────────────────────────────────
# Reason  : NDVI (vegetation index) depends on crop canopy structure
#           Rice paddy NDVI ≠ dryland Wheat NDVI
# Method  : Fill NaN with the median NDVI of the same crop

ndvi_miss = df["NDVI"].isna().sum()

for crop in df["Crop_Type"].unique():
    mask   = df["Crop_Type"] == crop
    median = df.loc[mask, "NDVI"].median()
    df.loc[mask & df["NDVI"].isna(), "NDVI"] = median

print(f"     Missing count : {ndvi_miss}")
print(f"     Method        : Median per Crop_Type")
print(f"     Remaining NaN : {df['NDVI'].isna().sum()}")
```

```python
# ── 3B : Bulk_Density_gcm3 — median per Soil_Type ───────────────────────────
# Reason  : Bulk density is a physical property of soil texture
#           Sandy soils have lower bulk density than Clay soils
# Method  : Fill NaN with the median of the same soil type

bd_miss = df["Bulk_Density_gcm3"].isna().sum()

for soil in df["Soil_Type"].unique():
    mask   = df["Soil_Type"] == soil
    median = df.loc[mask, "Bulk_Density_gcm3"].median()
    df.loc[mask & df["Bulk_Density_gcm3"].isna(), "Bulk_Density_gcm3"] = median

print(f"     Missing count : {bd_miss}")
print(f"     Remaining NaN : {df['Bulk_Density_gcm3'].isna().sum()}")
```

```python
# ── 3C : Sulfur, Zinc, Iron — median per (Crop_Type + Soil_Type) ────────────
# Reason  : Micronutrient availability depends on BOTH crop uptake AND soil
#           e.g. Zinc deficiency is common in Rice on clay soils
# Method  : Group by (Crop_Type, Soil_Type) → fill with group median
#           Fallback: overall median if group has no non-null values

micro_cols = ["Sulfur_kgha", "Zinc_ppm", "Iron_ppm"]
miss_before = {col: df[col].isna().sum() for col in micro_cols}

for col in micro_cols:
    group_median = df.groupby(["Crop_Type", "Soil_Type"])[col].transform("median")
    df[col] = df[col].fillna(group_median)           # group median first
    df[col] = df[col].fillna(df[col].median())       # overall fallback
    df[col] = df[col].round(2)

print(f"\n   Total missing values after imputation : {df.isnull().sum().sum()}")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 4 — OUTLIER DETECTION AND TREATMENT (WINSORIZATION)
# ─────────────────────────────────────────────────────────────────────────────

# Environmental and soil numerical columns
num_cols = [
    "Avg_Temp_C", "Max_Temp_C", "Min_Temp_C",
    "Rainfall_mm", "Humidity_Pct", "Solar_Radiation_MJm2",
    "Wind_Speed_kmh", "ET0_mm_day", "Water_Stress_Index",
    "Soil_pH", "Soil_Moisture_Pct", "Organic_Carbon_Pct",
    "Bulk_Density_gcm3", "N_kgha", "P_kgha", "K_kgha",
    "Sulfur_kgha", "Zinc_ppm", "Iron_ppm", "NDVI"
]

total_outliers = 0

for col in num_cols:
    Q1  = df[col].quantile(0.25)
    Q3  = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_fence = Q1 - 3 * IQR
    upper_fence = Q3 + 3 * IQR

    n_outliers = ((df[col] < lower_fence) | (df[col] > upper_fence)).sum()
    df[col]    = df[col].clip(lower=lower_fence, upper=upper_fence)

    if n_outliers > 0:
        print(f"   {col:<28} {n_outliers:>9,}  {lower_fence:>12.2f}  {upper_fence:>12.2f}")
    total_outliers += n_outliers

# Crop_Yield_kg_ha → per-crop IQR (IMPORTANT — crops have very different scales)

for crop in df["Crop_Type"].unique():
    mask = df["Crop_Type"] == crop
    Q1   = df.loc[mask, "Crop_Yield_kg_ha"].quantile(0.25)
    Q3   = df.loc[mask, "Crop_Yield_kg_ha"].quantile(0.75)
    IQR  = Q3 - Q1
    lo   = Q1 - 3 * IQR
    hi   = Q3 + 3 * IQR

    n_out = ((df.loc[mask, "Crop_Yield_kg_ha"] < lo) |
             (df.loc[mask, "Crop_Yield_kg_ha"] > hi)).sum()

    df.loc[mask, "Crop_Yield_kg_ha"] = \
        df.loc[mask, "Crop_Yield_kg_ha"].clip(lower=lo, upper=hi)

    print(f"   {crop:<14} {n_out:>9,}  {lo:>12.0f}  {hi:>12.0f}")
    total_outliers += n_out

print(f"\n   Total outliers winsorized : {total_outliers:,}")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 5 — DATA TYPE AND FORMAT STANDARDISATION
# ─────────────────────────────────────────────────────────────────────────────
# Ensure Year is integer (no decimals like 2015.0)
df["Year"] = df["Year"].astype(int)

# Standardise all categorical text: strip spaces + Title Case
cat_cols = [
    "Season", "Crop_Type", "Soil_Type", "Irrigation_Method",
    "Fertilizer_Type", "Pest_Incidence", "Disease_Incidence", "Yield_Category"
]

for col in cat_cols:
    df[col] = df[col].astype(str).str.strip().str.title()

# Pest and Disease: after .str.title() NaN becomes NaN again — fill once more
df["Pest_Incidence"]   = df["Pest_Incidence"].fillna("None")
df["Disease_Incidence"] = df["Disease_Incidence"].fillna("None")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 6 — FINAL VALIDATION (confirm everything is clean)
# ─────────────────────────────────────────────────────────────────────────────

checks = {
    "No missing values"                    : df.isnull().sum().sum() == 0,
    "No duplicate rows"                    : df.duplicated().sum() == 0,
    "No negative Elevation_m"              : (df["Elevation_m"] >= 0).all(),
    "Soil_pH in valid range (4 – 10)"      : df["Soil_pH"].between(4, 10).all(),
    "Humidity_Pct in valid range (0 – 100)": df["Humidity_Pct"].between(0, 100).all(),
    "NDVI in valid range (0 – 1)"          : df["NDVI"].between(0, 1).all(),
    "Water_Stress_Index in range (0 – 1)"  : df["Water_Stress_Index"].between(0, 1).all(),
    "All Crop_Yield_kg_ha > 0"             : (df["Crop_Yield_kg_ha"] > 0).all(),
    "Yield_Category classes balanced"      : df["Yield_Category"].value_counts().min() > 1000,
    "No zeros in Iron_ppm"                 : (df["Iron_ppm"] > 0).all(),
    "No zeros in Zinc_ppm"                 : (df["Zinc_ppm"] > 0).all(),
    "100% rows retained"                   : len(df) == 220000,
}

all_pass = True

for check in checks:
    if checks[check]:
        print("PASS", check)
    else:
        print("FAIL", check)
        all_pass = False

if all_pass:
    print("\nOverall: ALL CHECKS PASSED")
else:
    print("\nOverall: SOME CHECKS FAILED")
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# STEP 7 — BEFORE vs AFTER SUMMARY REPORT
# ─────────────────────────────────────────────────────────────────────────────
summary = [
    ("Total Rows",               "2,20,000",              str(len(df))),
    ("Duplicate Rows",           "0",                     "0"),
    ("Total Missing Values",     "1,53,955",              str(df.isnull().sum().sum())),
    ("Negative Elevation Rows",  "5,504",                 "0  (abs() applied)"),
    ("Zero Iron_ppm Rows",       "4,749",                 "0  (→ NaN → imputed)"),
    ("Zero Zinc_ppm Rows",       "4,892",                 "0  (→ NaN → imputed)"),
    ("Yield_Category 'Low'",     "2,06,407  (93.8%)",     f"{(df['Yield_Category']=='Low').sum():,}  (33%)"),
    ("Yield_Category 'Medium'",  "13,584   (6.2%)",       f"{(df['Yield_Category']=='Medium').sum():,}  (33%)"),
    ("Yield_Category 'High'",    "9        (0.004%)",     f"{(df['Yield_Category']=='High').sum():,}  (34%)"),
    ("Pest_Incidence NaN",       "55,044",                "0  (filled as 'None')"),
    ("Disease_Incidence NaN",    "54,961",                "0  (filled as 'None')"),
    ("Outliers Winsorized",      "—",                     f"{total_outliers}  (3×IQR capped)"),
]

print(f"\n   {'Metric':<32} {'BEFORE':>22}  {'AFTER':>30}")
for metric, before, after in summary:
    print(f"   {metric:<32} {before:>22}  {after:>30}")
```

```python
 # ─────────────────────────────────────────────────────────────────────────────
 # STEP 8 — POST-CLEANING STATISTICS
 # ─────────────────────────────────────────────────────────────────────────────

key_cols = [
    "Avg_Temp_C", "Rainfall_mm", "Humidity_Pct",
    "Soil_pH", "N_kgha", "P_kgha", "K_kgha",
    "Sulfur_kgha", "Zinc_ppm", "Iron_ppm",
    "NDVI", "Crop_Yield_kg_ha"
]
 
print(f"\n{df[key_cols].describe().round(2).to_string()}")
 
print(f"\n   Yield_Category distribution (cleaned):")
print(f"   {df['Yield_Category'].value_counts().to_dict()}")
 
print(f"\n   Crop-wise mean yield after cleaning:")
print(df.groupby("Crop_Type")["Crop_Yield_kg_ha"].mean().round(0).sort_values(ascending=False).to_string())
```

```python
# ─────────────────────────────────────────────────────────────────────────────
# SAVE CLEANED DATASET
# ─────────────────────────────────────────────────────────────────────────────
df.to_csv("crop_yield_cleaned_dataset.csv", index=False)

print(f"✅ Cleaned dataset saved → crop_yield_cleaned_dataset.csv")
print(f"   Final shape : {df.shape[0]:,} rows × {df.shape[1]} columns")
```
