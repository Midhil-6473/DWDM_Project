```python
import pandas as pd
import numpy as np

df = pd.read_csv("crop_yield_raw_dataset.csv")
df
```

```python
print(f"\n[2D] Yield_Category Correction (Per-Crop Percentile)")
print(f"     BEFORE:")
print(f"     {df['Yield_Category'].value_counts().to_dict()}")
 
df["Yield_Category"] = "Medium"
 
for crop in df["Crop_Type"].unique():
    mask   = df["Crop_Type"] == crop
    yields = df.loc[mask, "Crop_Yield_kg_ha"]
    p33    = yields.quantile(0.33)
    p66    = yields.quantile(0.66)
 
    df.loc[mask & (yields <= p33), "Yield_Category"] = "Low"
    df.loc[mask & (yields >  p66), "Yield_Category"] = "High"
 
    print(f"     {crop:<12} → Low ≤ {p33:.0f}  |  Medium  |  High > {p66:.0f} kg/ha")
 
print(f"\n     AFTER:")
print(f"     {df['Yield_Category'].value_counts().to_dict()}")
```
